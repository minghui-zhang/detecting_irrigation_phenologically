```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
raw = pd.read_csv('springboard_capstone1_trainingpts.csv')
```

```python
# eliminate system:index and .geo columns; these are the result of exporting data from Google Earth Engine

raw = raw.drop(['system:index', '.geo'], axis = 'columns')

raw.head()
```

```python
# the columns mod_B0_0 to mod_B0_91 represent the timeseries of unsmoothed EVI values, spaced every 8 days
mod_ts_cols = [colname for colname in raw if colname.startswith('mod')]

# the order of the mod_ts_cols needs to be shuffled into chronological order before they can be interpolated
# first, create a list of integers representing the order to sort by
sort_order = [int(colname.split('_')[2]) for colname in mod_ts_cols]

# sort the names of the timeseries columns
sorted_mod_ts_cols = [x for _,x in sorted(zip(sort_order, mod_ts_cols))]

# get the timeseries data before interpolation
uninterpolated_ts = raw[sorted_mod_ts_cols]

# get the timeseries data and linearly interpolate across the columns
interpolated_ts = raw[sorted_mod_ts_cols].apply(lambda x: x.interpolate(), axis = 1)

# choose an observation with a missing value, plot it with and without interpolation
unint_pt = uninterpolated_ts[uninterpolated_ts.isnull().any(axis = 1)].loc[28, :]
int_pt = interpolated_ts.loc[28,:]
```

```python
# this plot confirms that linear interpolation was done correctly
plt.scatter(list(range(0, 91)), int_pt, label = 'interpolated')
plt.scatter(list(range(0, 91)), unint_pt, label = 'raw')
plt.title('EVI timeseries')
plt.xlabel('MODIS acquisition time')
plt.ylabel('EVI')
plt.legend()
plt.show()
```

```python
# handle missing values for the non-timeseries features. these can't be linearly interpolated
non_ts_cols = list(set(raw.columns) - set(mod_ts_cols))

# should we get rid of entire columns if they have too much missing data (i.e. more than 2% of total observations)
cols_with_enough = [colname for colname in non_ts_cols if raw[colname].count()/len(raw) > 0.98]
cols_without_enough = list(set(non_ts_cols) - set(cols_with_enough))
print(cols_without_enough) # we find that only EVI values at certain phenological stages are commonly NaN

#raw[non_ts_cols].info()
```

```python
# combine the timeseries and non-timeseries dataframes before eliminating more rows
joined = pd.concat([interpolated_ts, raw[non_ts_cols]], axis = 1)
```

```python
# get rid of rows that are completely null in the phenospectral features. it's unlikely that classification will succeed without phenospectral information
phenospectral_cols = list(set(non_ts_cols) - set(['elevation', 'lat', 'lon', 'state', 
                                                  'hillshade', 'region', 'aspect', 'slope',
                                                 'class']))
joined.dropna(subset = phenospectral_cols, how = 'all', inplace = True)

print(len(joined)) # 11793 out of 12000 points are retained
```

```python
# because missing-ness in phenological features is likely meaningful, we don't impute its value by the class mean
# instead, we add a new column to denote missing data for each phenological feature

for colname in phenospectral_cols:
    joined[colname + '_missing'] = joined[colname].isnull()
```

```python
# outliers

# set reasonable limits for each phenological parameter
reasonable_limits = {'peakDay_second': [60, 365], 'fittedAmplitude_first': [0, 2],
                    'EVI_at_peak1': [0.1, 3], 'EVI_at_min1': [0, 0.5],
                    'maxEVI_first': [0.1, 3], 'EVI_at_min2': [0, 0.5],
                    'EVI_at_infl2': [0.1, 3], 'quarterPeriod_first': [10, 50],
                    'EVI_at_peak2': [0.1, 3], 'fittedAmplitude_second': [0, 2], 
                    'quarterPeriod_second': [10, 50], 'peakDiff': [0, 300],
                    'peakDay_first': [0, 200], 'EVI_at_infl1': [0.1, 3],
                    'maxEVI_second': [0.1, 3]}

for colname in phenospectral_cols:
    #joined[colname].plot(kind = 'box')
    #plt.ylabel(colname)
    lower_limit = reasonable_limits[colname][0]
    upper_limit = reasonable_limits[colname][1]
    sns.violinplot(joined[colname])
    plt.axvline(lower_limit, color = 'red')
    plt.axvline(upper_limit, color = 'red')
    plt.title(colname)
    plt.show()
```

Peakday_second has a bimodal distribution; one at a reasonable range, and another around 500. since the peak day should occur completely within the year, the peak day should never exceed 365 the way it does here. However, note that a phenological estimate that's outside of reasonable bounds doesn't necessarily mean it's a bad data point; it just means that the real EVI timeseries doesn't fit into initial assumptions of how the EVI should be evolving, which is based on rainfed agriculture. The fact that a phenological parameter lies out of reasonable bounds may mean that it's more likely to the irrigation (and therefore doesn't follow our rainfed assumptions). The quarterperiod and peakday, when out of bounds, may still contain useful information.

However, out-of-bounds values in other phenological estimates may be cause for concern. For example, EVI for regular agricultural crops would not exceed 3. We see from the violin plot that EVI_at_peak1 does exceed 3 for at least one outlier point. It follows that the fitted amplitude of the EVI curve should not exceed 3; this outlier is observed in fittedAmplitude_first. When EVI values exceed what is physically possible, the underlying satellite data is likely in error.

```python
joined['EVI_at_peak1'].plot(kind = 'box')
plt.title('EVI_at_peak1')
plt.ylabel('EVI')
plt.show()

joined['fittedAmplitude_first'].plot(kind = 'box')
plt.title('fittedAmplitude_first')
plt.ylabel('EVI amplitude')
plt.show()
```

The boxplots show only one outlier point for EVI_at_peak_1 and only five outlier points for fittedAmplitude_first. We eliminate these points.

```python
joined = joined[(joined.EVI_at_peak1 <= 3) & (joined.fittedAmplitude_first <= 3)]

joined.info()
```

```python
# export cleaned data
joined.to_csv('trainingpts_cleaned.csv')
```

```python
# data cleaning workflow for prediction points: repeat the same thing as for training data
year = 2004
prediction_raw = pd.read_csv('springboard_capstone1_testpts_' + str(year) + '.csv')
prediction_raw = prediction_raw.drop(['system:index', '.geo'], axis = 'columns')

# the columns mod_B0_0 to mod_B0_91 represent the timeseries of unsmoothed EVI values, spaced every 8 days
mod_ts_cols = [colname for colname in prediction_raw if colname.startswith('mod')]

# the order of the mod_ts_cols needs to be shuffled into chronological order before they can be interpolated
# first, create a list of integers representing the order to sort by
sort_order = [int(colname.split('_')[2]) for colname in mod_ts_cols]

# sort the names of the timeseries columns
sorted_mod_ts_cols = [x for _,x in sorted(zip(sort_order, mod_ts_cols))]

# get the timeseries data and linearly interpolate across the columns
prediction_interpolated_ts = prediction_raw[sorted_mod_ts_cols].apply(lambda x: x.interpolate(), axis = 1)

# combine the timeseries and non-timeseries dataframes before eliminating more rows
prediction_joined = pd.concat([prediction_interpolated_ts, prediction_raw[non_ts_cols]], axis = 1)

# handle missing values
prediction_joined.dropna(subset = phenospectral_cols, how = 'all', inplace = True)
for colname in phenospectral_cols:
    prediction_joined[colname + '_missing'] = joined[colname].isnull()

# exclude outliers
prediction_joined = prediction_joined[(prediction_joined.EVI_at_peak1 <= 3) & (prediction_joined.fittedAmplitude_first <= 3)]
    
# export cleaned data
prediction_joined.to_csv('predictionpts_cleaned_' + str(year) + '.csv')
```
